```python
import matplotlib.pyplot as plt
x="density"
y="area"
#c=['r','y','g','b']
plt.scatter(x, y,data=df,)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Assuming you want to create a histogram for the 'country_code' variable
#c=['r','y','g','b']
sns.histplot(df['growthRate'])

# Customize the plot
plt.title('Histogram of growthRate')
plt.xlabel('growthRate')
# Show the plot
plt.show()

```

```python
import matplotlib.pyplot as plt

# Assuming you want to create a histogram for the 'country_code' variable
sns.barplot(df['growthRate'] ,color="skyblue",edgecolor="black")

# Customize the plot
plt.title('bar of growthRate')
plt.xlabel('growthRate')
# Show the plot
plt.show()
```

```python
x="growthRate"
y="area"
sns.lineplot(data=df, x="growthRate", y="area")
plt.show
```

# Bar color demo

```python

import matplotlib.pyplot as plt

fig, ax = plt.subplots()

growthRate = ['0.0000', '0.0068', '0.0038', '0.0064']
density = [  151.2926, 476.6507, 36.9820, 146.7369]
bar_labels = ['red', 'blue', '_red', 'orange']
bar_colors = ['tab:red', 'tab:blue', 'tab:red', 'tab:orange']

ax.bar(growthRate, density, label=bar_labels, color=bar_colors)

ax.set_ylabel('growthRate')
ax.set_title('growthRate by kind and color')
ax.legend(title='color')

plt.show()
```
